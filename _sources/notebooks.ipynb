```python
dataset_url="https://raw.githubusercontent.com/Theresia028/datamining/gh-pages/ds_salaries.csv"
data=pd.read_csv(dataset_url, index_col=0)
```

```python
data
```

```python
def diss(x,y):
  a = 2
  b = 0
  for col in ['salary','salary_in_usd']:
    if data[col][x] == data[col][y]:
      b+=1
    return (a-b)/a
```

Menghitung Dissimilarity

```python
print(diss(1,1))
print(diss(1,2))
print(diss(3,3))
```
