```python
import numpy as np
import pandas as pd
```

```python
dataset_url="https://raw.githubusercontent.com/Theresia028/datamining/gh-pages/ds_salaries.csv"
data=pd.read_csv(dataset_url, index_col=0)
```

```python
data
```

```python
def diss(x,y):
  a = 2
  b = 0
  for col in ['salary','salary_in_usd']:
    if data[col][x] == data[col][y]:
      b+=1
    return (a-b)/a
```

Menghitung Dissimilarity

```python
print(diss(1,1))
print(diss(1,2))
print(diss(3,3))
```

```python
import numpy as np
import pandas as pd
```

```python
from sklearn import datasets
iris = datasets.load_iris()
x = iris.data[:,:1] # we only take the first two features.
y = iris.target
```

```python
x
```

```python
df = pd.DataFrame(x,columns = ['sepal'])
```

```python
df
```

```python
from sklearn.preprocessing import KBinsDiscretizer
est = KBinsDiscretizer(n_bins=4, encode='ordinal', strategy='uniform')
df['sepal'] = est.fit_transform(df[['sepal']])
```

```python
df.to_csv('out.csv')
```
